```python
def add_field_to_df(path_df, sport_name, col_name='Sport'): 
    df_sport = pd.read_csv(path_df, on_bad_lines='skip')
    df_sport[col_name] = [sport_name]*df_sport.shape[0]
    return df_sport
```

```python
import os
directory = './fake-dataset'
for root, dirs, files in os.walk(directory):
    for file in files:
        file_path = os.path.join(root, file)
        
        sport_name = file.split('-')[0]
        print(file_path, sport_name)
        df_res = add_field_to_df(file_path, sport_name, 'Sport')
        df_res.to_csv(file_path, index=False)
```

```python
## get a df of all products
df_products = pd.DataFrame()

directory = './fake-dataset'
for root, dirs, files in os.walk(directory):
    for file in files:
        df_ = pd.read_csv(os.path.join(root, file))
        df_products = pd.concat([df_products, df_])
```

```python
df_products.columns = df_products.columns.str.strip()

for col in df_products.columns: 
    df_products[col] = df_products[col].str.strip()

df_products['Product ID'] = ["PID"+"{:03}".format(num) for num in np.arange(0, 66, 1)]
```

```python
# # Function to randomly select a product
# def random_product_selection(df_products):
#     return df_products.sample(n=1).iloc[0]

# # Randomly select a product for each customer
# random_products = [random_product_selection(df_products) for _ in range(len(df_customers))]

# # Convert the list of Series to a DataFrame
# random_products_df = pd.DataFrame(random_products).reset_index(drop=True)

# # Concatenate the customer DataFrame with the randomly selected products
# df_sales = pd.concat([df_customers, random_products_df], axis=1)

# Function to randomly select unique products for a customer
def random_unique_products(df_products, num_products):
    return df_products.sample(n=num_products).reset_index(drop=True)

# Expanded customer records
expanded_records = []

# Iterate through each customer and randomly create multiple records
for index, customer in df_customers.iterrows():
    # Randomly decide how many records to create for this customer
    # num_records = np.random.randint(1, 6)  # At least 1 record, at most 5 products
    
    # Weighted random choice: 80% chance of 1 record, 20% chance of more than 1 record
    num_records = np.random.choice(
        [1, np.random.randint(2, 10)],
        p=[0.8, 0.2]
    )
    selected_products = random_unique_products(df_products, num_records)
    
    for _, product in selected_products.iterrows():
        expanded_record = customer.to_dict()
        expanded_record.update(product.to_dict())
        expanded_records.append(expanded_record)

# Create the final expanded DataFrame
df_sales = pd.DataFrame(expanded_records)
```
